```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1)
y = np.linspace(0, 1)

X ,Y = np.meshgrid(x, y)

Z = np.sqrt(X**2 + Y**2)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
from scipy import integrate
f = lambda x, y: x + y
x_initial = -1
x_final = 1
y_initial = lambda x: 0
y_final = lambda y: 1
integrate.dblquad(f, x_initial, x_final, y_initial, y_final)[0]

f = lambda x: x
integrate.quad(f, 0, 1)
```

```python
def f(x):
    y = np.zeros_like(x)
    y = -x
    y[(-1 < x) & (x < 1)] -= 3
    return y

x = np.linspace(-5, 5, 1000)
plt.plot(f(x), x)
```
